```python
df = pd.read_csv("C:/Users/mford.MSBIOWORKS/Project_1/state_counts.csv")
# print(monthly_energy_production_df)
df.head(5)
```

```python
df.sort_values(by='State', ascending=True, inplace = True)
df.head(5)
```

```python
# row_count = len(df)
# print("Number of rows in the DataFrame:", row_count)
print(df)
```

```python

# Calculate the total count for all states
total_count_all_states = df['Count'].sum()

# Calculate the percentage for each row based on the total count for all states
df['Percentage'] = (df['Count'] / total_count_all_states) * 100

# Display the resulting data frame with percentage values
print(df)
```

```python
# Merge the GeoDataFrame with your data based on a common column (e.g., 'State' column)
merged_data = gdf.merge(df, left_on='NAME', right_on='State', how='left')
```

```python
row_count = len(merged_data)
print("Number of rows in the DataFrame:", row_count)
```

```python
merged_data = merged_data.to_crs(epsg=3395)

fig, ax = plt.subplots(1, 1, figsize=(12, 8))
merged_data.plot(column='Percentage', cmap='viridis_r', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

for x, y, label in zip(merged_data.geometry.centroid.x, merged_data.geometry.centroid.y, merged_data['Percentage']):
    ax.text(x, y, f'{label:.1f}', fontsize=8, ha='center', va='center', color='white')


plt.title('Choropleth Map of US States')
plt.show()



# # Assuming you have a GeoDataFrame 'merged_data' with geometry, 'Percentage', and 'Count' columns
# # Reproject to an appropriate projected CRS (e.g., EPSG:3395 for world maps)
# merged_data = merged_data.to_crs(epsg=3395)

# fig, ax = plt.subplots(1, 1, figsize=(12, 8))
# merged_data.plot(column='Percentage', cmap='viridis_r', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

# for x, y, label in zip(merged_data.geometry.centroid.x, merged_data.geometry.centroid.y, merged_data['Percentage']):
#     ax.text(x, y, f'{label:.1f}', fontsize=8, ha='center', va='center', color='white')  # Format the label to one decimal place

# plt.title('Choropleth Map of US States')
# plt.show()



















########################################################################

# # Replace 'gdf' with your GeoDataFrame and 'value_column' with your quantitative value column
# log_values = np.log1p(merged_data['Count'])  # Apply logarithm to your data

# # Create a logarithmic color scale
# norm = LogNorm(vmin=log_values.min(), vmax=log_values.max())
# cmap = plt.get_cmap('YlGnBu')

# # Plot the GeoDataFrame with a logarithmic color scale
# fig, ax = plt.subplots(figsize=(12, 8))
# gdf.plot(column=log_values, cmap=cmap, linewidth=0.8, ax=ax, edgecolor='0.8', norm=norm, legend=True)
# plt.title('Heatmap with Logarithmic Color Scale')
# plt.show()


##########################################################################


# # Replace 'gdf' with your GeoDataFrame and 'value_column' with your quantitative value column
# log_values_base_2 = np.log2(merged_data['Count'] + 1)  # Apply logarithm to base 2 to your data

# # Create a logarithmic color scale
# vmin = log_values_base_2.min()
# vmax = log_values_base_2.max()
# norm = LogNorm(vmin=vmin, vmax=vmax)
# cmap = plt.get_cmap('YlGnBu')

# # Plot the GeoDataFrame with a logarithmic color scale to base 2
# fig, ax = plt.subplots(figsize=(12, 8))
# merged_data.plot(column=log_values_base_2, cmap=cmap, linewidth=0.8, ax=ax, edgecolor='0.8', legend=True, norm=norm)

# # Customize the legend label
# cax = fig.get_axes()[1]
# cax.set_ylabel("Log2 Scale")

# plt.title('Heatmap with Logarithmic (Base 2) Color Scale')
# plt.show()




```

```python
# Assuming you have a GeoDataFrame 'merged_data' with geometry, 'Percentage', and 'Count' columns
# Reproject to an appropriate projected CRS (e.g., EPSG:3395 for world maps)
merged_data = merged_data.to_crs(epsg=3395)

fig, ax = plt.subplots(1, 1, figsize=(12, 8))
merged_data.plot(column='Percentage', cmap='viridis_r', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

for x, y, label in zip(merged_data.geometry.centroid.x, merged_data.geometry.centroid.y, merged_data['Percentage']):
    if merged_data.loc[merged_data['State'] == 'Texas', 'Percentage'].values[0] == label:
        ax.text(x, y, f'{label:.1f}%', fontsize=6, ha='center', va='center', color='white')  # Texas label in black
    else:
        ax.text(x, y, f'{label:.1f}%', fontsize=6, ha='center', va='center', color='black')  # Other labels in white

plt.title('Choropleth Map of US States')
plt.show()
```

```python
# Assuming you have a GeoDataFrame 'merged_data' with geometry, 'Percentage', and 'Count' columns
# Reproject to an appropriate projected CRS (e.g., EPSG:3395 for world maps)
merged_data = merged_data.to_crs(epsg=3395)

fig, ax = plt.subplots(1, 1, figsize=(12, 8))
merged_data.plot(column='Percentage', cmap='viridis_r', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

for x, y, label in zip(merged_data.geometry.centroid.x, merged_data.geometry.centroid.y, merged_data['Percentage']):
    if label >= 0.1:
        ax.text(x, y, f'{label:.1f}%', fontsize=6, ha='center', va='center', color='black' if label != 1.0 else 'white')
    else:
        ax.text(x, y, '', fontsize=6, ha='center', va='center')  # Hide labels for values less than 0.1%

plt.title('Choropleth Map of US States')
plt.show()
```
